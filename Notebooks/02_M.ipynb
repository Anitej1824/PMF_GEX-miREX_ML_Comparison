# 02 - Metadata Processing & Label Construction

This notebook processes sample metadata from GEO (GSE53482)
and constructs labels for downstream machine learning tasks.

Key steps:
- Load processed gene expression data
- Extract sample metadata from GEO series matrix file
- Parse metadata into structured format
- Construct classification labels (PMF vs Control)
- Align metadata with expression dataset
- Validate dataset integrity
- Save processed metadata

```python
# Loading Required Libraries
import pandas as pd
import sys
import os

sys.path.append(os.path.abspath(".."))

# Import custom functions for data loading and preprocessing
from Src.Data_Preprocessing import extract_metadata, parse_metadata
```

```python
gene_file = "../Data/Raw/GSE53482-GPL13667_series_matrix.txt.gz"
```

```python
meta_gene_raw = extract_metadata(gene_file)
meta_gene_clean = parse_metadata(meta_gene_raw)

meta_gene_clean.head()
```

```python
meta_gene_clean["disease"].value_counts()
```

```python
meta_gene_clean["label"] = meta_gene_clean["disease"].map({
    "PMF": 1,
    "PB CTR": 0,
    "BM CTR": 0
})

meta_gene_clean["label"].value_counts()
```

```python
meta_gene_clean.index = meta_gene_clean.index.str.replace('"', '')
```

```python
gene_df = pd.read_csv("../Data/Processed/gene_expression.csv", index_col=0)
```

```python
gene_df = gene_df.loc[meta_gene_clean.index]
assert gene_df.index.equals(meta_gene_clean.index)
```

```python
gene_df.to_csv("../Data/Processed/gene_expression_aligned.csv")
meta_gene_clean.to_csv("../Data/Processed/meta_gene_data.csv")
```
