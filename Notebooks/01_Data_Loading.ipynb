# 01 - Data Loading (Gene Expression + miRNA Expression)

This notebook loads transcriptomic datasets from GEO (GSE53482),
including both gene expression (GEP) and miRNA expression (miEP) profiles.

The data is derived from CD34+ cells of PMF patients and healthy controls.

Steps:
- Load GEO series matrix files (gene + miRNA)
- Extract expression tables
- Clean and format data (remove quotes, set indices, convert to numeric)
- Transpose data (samples as rows, features as columns)
- Save processed datasets for downstream analysis

### Import Custom Data Processing Functions

```python
# Loading Required Libraries
import sys
import os

sys.path.append(os.path.abspath(".."))

# Import custom functions for data loading and preprocessing
from Src.Data_Preprocessing import load_geo_matrix, clean_expression_data
```

### Define Data File Paths

Specify the file paths for the raw GEO datasets:
- Gene expression (GEP)
- miRNA expression (miEP)

```python
# File paths for raw GEO datasets
gene_file = "..\Data\Raw\GSE53482-GPL13667_series_matrix.txt.gz"
mirna_file = "..\Data\Raw\GSE53482-GPL14613_series_matrix.txt.gz"
```

### Load Gene Expression Data

Load the gene expression dataset from the GEO series matrix file.

```python
# Load raw gene expression data
gene_df = load_geo_matrix(gene_file)

# Check dimensions (rows = probes, columns = samples)
gene_df.shape
```

```python
# Preview raw gene expression data
gene_df.head()
```

### Clean and Transform Gene Expression Data

Apply preprocessing steps to convert the raw Gene Expression data into an ML-ready format.

This includes:
- Removing formatting artifacts (quotation marks)
- Setting probe IDs as index
- Converting values to numeric
- Transposing the matrix


```python
# Clean and transform gene expression data
gene_df = clean_expression_data(gene_df)

# Check new shape (rows = samples, columns = genes)
gene_df.shape
```

```python
# Preview transformed gene expression data
print("Gene expression shape:", gene_df.shape)
```

### Verify Gene Expression Dataset Dimensions

Confirm the number of samples and features after preprocessing.

```python
# Display number of samples and genes
print("Samples:", gene_df.shape[0])
print("Genes:", gene_df.shape[1])
```

### Load miRNA Expression Data

Load the miRNA expression dataset from the GEO series matrix file.

```python
# Load raw miRNA expression data
mirna_df = load_geo_matrix(mirna_file)

# Check dimensions (rows = probes, columns = samples)
mirna_df.shape
```

```python
# Preview raw miRNA expression data
mirna_df.head()
```

### Clean and Transform miRNA Expression Data

Apply preprocessing steps to convert the raw miRNA Expression data into an ML-ready format. (Follow same steps as for the Gene Expression dataset)

```python
# Clean and transform miRNA expression data
mirna_df = clean_expression_data(mirna_df)

# Check new shape (rows = samples, columns = genes)
mirna_df.shape
```

```python
# Preview transformed gene expression data
mirna_df.head()
```

### Verify miRNA Expression Dataset Dimensions

Confirm the number of samples and features after preprocessing.

```python
# Display number of samples and miRNAs
print("Samples:", mirna_df.shape[0])
print("miRNAs:", mirna_df.shape[1])
```

### Save Processed Data

Save cleaned datasets for reuse in downstream analysis notebooks.

```python
# Create processed directory if it doesn't exist
os.makedirs("../Data/Processed", exist_ok=True)

# Save processed data
gene_df.to_csv("../Data/Processed/gene_expression.csv")
mirna_df.to_csv("../Data/Processed/mirna_expression.csv")
```
